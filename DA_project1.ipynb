```python
import requests
```

```python
data=requests.get("https://visaguide.world/africa/")
data
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(data.text,"html.parser")
soup
```

```python
dt = soup.find_all("table")
dt
```

```python
dt[0]
```

```python
dt[1]
```

```python
headers=dt[1].find_all("th")
headers=[i.text for i in headers]
headers
```

```python
rows=dt[1].find_all("td")
rows=[i.text for i in rows]
rows
```

```python
d={}
for i in headers:
  d[i]=[]
d
```

```python
for i in range(len(rows)):
  if i % 4==0:
    d['Country'].append(rows[i])
  if i % 4==1:
    d['Capital'].append(rows[i])
  if i % 4==2:
    d['Area km2'].append(rows[i])
  if i % 4==3:
    d['Population'].append(rows[i])
```

```python
import pandas as pd
df =pd.DataFrame(d)
df
```

```python
df['Area km2']=df['Area km2'].str.replace(",",'').astype(int)
```

```python
df.sort_values(by='Area km2',ascending=False,inplace=True)
df
```

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(50,20))
x=df['Country']
y=df['Area km2']
plt.xticks(rotation=90, fontsize=40)
plt.yticks(fontsize=40)
plt.bar(x,y)

plt.show()
```
